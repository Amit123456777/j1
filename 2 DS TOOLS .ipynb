```python
import pandas as pd
df = pd.read_csv("nba.csv")
df
```

```python
df.sort_values("Name", axis=0, ascending = True, inplace = True, na_position = 'last')
df
```

```python
df.sort_values("Name", axis=0, ascending = False, inplace = True, na_position = 'last')
df
```

```python
df.sort_values("Age", axis=0, ascending = True, inplace = True, na_position = 'last')
df
```

```python
df.sort_values("Age", axis=0, ascending = False, inplace = True, na_position = 'last')
df
```
