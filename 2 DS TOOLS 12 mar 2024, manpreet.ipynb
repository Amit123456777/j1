```python
import pandas as pd
df = pd.DataFrame({"name": ["b","c","a","e","d","f","g","h","i","j"],
                   "age": [10,20,30,40,10,30,44,21,32,22],
                   "salary": [1000,4000,2000,3000,7000,5000,6000,10000,9000,8000],
                   "departments": ["sale","purchase", "employees", "hr", "staff", "student", "teacher","office", "college", "bank"]
                  },
                   index= ["row1", "row2", "row3", "row4", "row5","row6", "row7", "row8", "row9", "row10"]
                                   )
df
```

```python
# que 4
df.sort_values('age',axis=0,ascending=False,inplace=True,na_position='first')
df
```
