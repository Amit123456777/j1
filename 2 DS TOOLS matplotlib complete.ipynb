# BAR PLOT

```python
import matplotlib.pyplot as plt
x = ["python","c","c++","java"]
y = [85,70,60,82]
plt.bar(x,y)
plt.show()
# bar function
# show function
# x value can be name and numbers
```

```python
# label,  title and fontsize
import matplotlib.pyplot as plt
x = ["python","c","c++","Java"]
y = [85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("wscube", fontsize = 15)
plt.bar(x,y)
plt.show(x,y)
```

```python
# width and color
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
plt.bar(x,y,width=0.4,color="y")
plt.show()
```

```python
# multiple color
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [85,70,60,82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
c = ["y", "b", "m", "g"] # m magenta
plt.bar(x,y,width=0.4,color = c)
plt.show()
```

```python
# align = "edge"
# align = "center" (by default)
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [80,70,60,82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
c = ["y", "b", "m", "g"]
plt.bar(x,y,width=0.4,color = c,align="edge")
plt.show()
```

```python
# edgecolor
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [80, 70, 60, 82]
plt.xlabel("language", fontsize= 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
c = ["y", "b", "m", "g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r")
plt.show()
```

```python
# linewidth
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [80, 70, 60, 82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
c = ["y", "b", 'm', "g"]
plt.bar(x,y,width=0.4, color=c, edgecolor="r", linewidth=5)
plt.show()

# solid linewidth
```

```python
# linestyle

import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [80, 70, 60, 82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
c = ["y", "b", "m", "g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r", linewidth = 5, linestyle=":")
plt.show()

# : color gives doted linestyle
```

```python
# alplha -- for/givestransparency
# value lie (0 - 1) float value
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [80, 70, 60, 82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
c = ["y", "b", "m", "g"]
plt.bar(x, y, width = 0.4, color = c , edgecolor = "r", linewidth=5, linestyle = ":", alpha = 0.5)
plt.show()
     
```

```python
# label+legend
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [80,70,60,82]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
c = ["y", "b", "m", "g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r",label="wscube")
plt.legend()
plt.show()
```

# stack bar charts


```python

# how to make multiple bar graph
import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [85,70,60,82]
z = [20,30,40,50]
plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
plt.bar(x, y, width = 0.4, color = "r", label = "popularity")
plt.bar(x, z, width = 0.4, color = "y", label = "popularity1")
plt.legend()
plt.show()
```

```python
# above graph is overlaped es dusre ke upar.
# but we want to show side by side (es dusre ko pass me ek sath show krna hai)
# comparision bar graph
import matplotlib.pyplot as plt

import numpy as np

x = ["python", "c", "c++", "Java"]
y = [80 ,70,60,82]
z = [40,30,40,50]
width = 0.2

p = np.arange(len(x)) # make an array of x length
p1 = [j+width for j in p]

plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)

plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")

plt.xticks(p+width/2,x,rotation=20)   # tick- intervals 0to10, 10to20, 20to30
plt.legend() 
plt.show()


# legend centre se corner me bi kr skte hai-using location parameter
```

```python
# barh = horizontal bar graph

import matplotlib.pyplot as plt
x = ["python", "c", "c++", "Java"]
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)

plt.barh(x, y,color = "r", label = "popularity")
plt.bar(x, z,color = "y", label = "popularity1")

plt.legend()
plt.show()
```

```python
# syntax

import matplotlib.pyplot as plt
x = []
y = []
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
# title, label and its fontsize
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
plt.scatter(day,no)
plt.title("Scatter Plot", fontsize = 15)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.show()
```

```python
# color
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
plt.scatter(day,no,color="r")
plt.title("Scatter Plot", fontsize = 15)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("NO,", fontsize = 15)
plt.show()
```

```python
# multi/different color
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
colors = ["r","y","g","b","r","g","r"]
plt.scatter(day,no,c=colors)
plt.title("Scatter Plot", fontsize = 15)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.show()
```
