```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib.inline
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Rahul\\Downloads\\company_sales_data.csv")
df
```

```python
# question_1
plt.plot(df["month_number"],df["total_profit"])
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.show()
```

```python
# question_2

profitList = df ['total_profit']
monthList  = df ['month_number']

plt.plot(monthList, profitList, label = 'Profit data of last year', 
      color='r', marker='o', markerfacecolor='k', 
      linestyle='--', linewidth=3)
      
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')
plt.title('Company Sales data of last year')
plt.xticks(monthList)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

```python
x=df["month_number"]
labels=["facecream","facewash","toothpaste","bathingsoap",
        "shampoo","moisturizer"]
plt.plot(x,df["facecream"],x,df["facewash"],x,df["toothpaste"],
         x,df["bathingsoap"],x,df["shampoo"],x,df["moisturizer"])
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.legend(labels)
plt.show()
```

```python
# ques 4
plt.scatter(df["month_number"],df["toothpaste"])
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.legend("Toothpaste Sales Data")
plt.show()
```

```python
# question_5
x=df["month_number"]
w=0.2
legend=["Facecream sales data","facewash sales data"]
plt.bar(x,df["facecream"],w)
x2=[i+w for i in x]
plt.bar(x2,df["facewash"],w)
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.title("Facewash and Facecream sales data")
plt.legend(legend)
plt.show()
```

```python
# question_6
plt.bar(df["month_number"],df["bathingsoap"])
plt.title("Bathingsoap sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.grid()
plt.show()
```

```python
# question_7
plt.hist(df["total_profit"])
plt.title("Profit data")
plt.xlabel("Profit range in dollars")
plt.ylabel("Actual profit in dollar")
plt.legend("Profit data")
plt.show()
```

```python
# question_8
labels=["Facecream","Facewash","Toothpaste","Bathingsoap",
        "Shampoo","Moisturizer"]
data=[2900,1760,7400,14400,1800,1760]
plt.pie(data,labels=labels)
plt.legend(labels)
plt.show()
```

```python
# question_9
x=df["month_number"]
y1=df["bathingsoap"]
y2=df["facewash"]

plt.plot(x,y1,linewidth=3,color="k")
plt.scatter(x,y1,color='k')
plt.title("Sales data of a Bathingsoap")
plt.xlabel("Month Number")
plt.ylabel("Sales units in numbers")
plt.show()

plt.plot(x,y2,linewidth=3,color="r")
plt.scatter(x,y2,color='r')
plt.title("Sales data of a Facewash")
plt.xlabel("Month Number")
plt.ylabel("Sales units in numbers")
plt.show()
```

```python
# question_10
 
colors=["pink","b","r","k","g","y"]
plt.stackplot(x,df["facecream"],df["facewash"],df["toothpaste"]
         ,df["bathingsoap"],df["shampoo"],
              df["moisturizer"],colors=colors)
plt.title("All product sales data using stack plot")
plt.xlabel("Month Number")
plt.ylabel("Sales units in numbers")
plt.legend(labels)
plt.show()
```
