```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = [33,52,12,17,42,48]
plt.pie(data)
plt.show()

# (33/total)*360degree
# 33/total = is called propertion.
```

```python
# adding labels 
label = ["Python","c","c++","Php","Java","Ruby"]
plt.pie(data,labels=label)
plt.show()
```

```python
plt.figure(figsize = (9,7))
plt.pie(data)
plt.title("the result of the programming language survey :",fontsize = 15)
plt.legend(label)
plt.show()
```

# explode

```python
explode = [0.0,0.08,0.00,0.00,0.00,0.00]
plt.pie(data, explode = explode,labels = label)
plt.show()
```

```python
explode = [0.0,0.008,0.00,0.00,0.00,0.00]
plt.pie(data, explode = explode,labels = label)
plt.show()
```

```python
explode = [0.0,0.08,0.00,0.00,0.00,0.00]
plt.pie(data, explode = explode,labels = label)
plt.show()

# explode parameter is an array 
```

```python
# only 10 categories allowed
# (1/30)*360=12

# frequency of python = 12
```

```python
# colors
colors = ['saddlebrown','wheat','crimson','lightgrey','lightblue','darkblue']
explode = [0.0,0.08,0.00,0.00,0.00,0.00]
plt.pie(data, explode = explode,labels = label, colors = colors, shadow = True)
plt.show()
```
