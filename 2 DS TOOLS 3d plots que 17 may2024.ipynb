```python
# question 1. generate a 3d graph defining random points using a scatter plot.
```

```python
x = np.random.randn((100))
y = np.random.randn((100))
z = np.random.randn((100))

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
scatter = ax.scatter(x, y,z, marker = 'o')
```

```python
# question_2
# change the x ticks y ticks and z ticks as follows
# xticks = 100, 125, 150,175,200
# yticks = 20, 55, 90, 125, 160
# zticks = 5, 15,25,35
```

```python
from mpl_toolkits.mplot3d  import Axes3D
x = np.random.randn((100))
y = np.random.randn((100))
z = np.random.randn((100))


fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y,z, marker = 'o')
ax.set_xticks([100, 125, 150, 175, 200])
ax.set_yticks([20, 55, 90, 125, 160])
ax.set_zticks([5, 15, 25, 35])
plt.grid()
plt.show()
```
