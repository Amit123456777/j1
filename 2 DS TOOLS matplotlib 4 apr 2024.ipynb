```python
import matplotlib.pyplot as plt
x = ["python", "java", "c" ,"cpp"]
y = [100,400,200,300]
plt.bar(x,y, color= ["midnightblue"],edgecolor="k", linewidth = 5, linestyle = ":", alpha = 0.5)
plt.xlabel("subject")
plt.ylabel("marks")
plt.title("score")

plt.show()
```

```python
# add a label to the graph saying performance

import matplotlib.pyplot as plt
x = ["python", "java", "c" ,"cpp"]
y = [100,400,200,300]
plt.bar(x,y, color= ["midnightblue"],edgecolor="k", linewidth = 5, linestyle = ":", alpha = 0.5, label = "performance")
plt.legend()
plt.xlabel("subject")
plt.ylabel("marks")
plt.title("score")

plt.show()
```

```python
# scatter plot

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.xlabel("x axis",fontsize = 15)
plt.ylabel("y axis",fontsize = 15)
bold = [1500,400,6000,700,400]
plt.scatter(x,y, color ="k", s=bold, edgecolor = "k", marker = "*", linewidth=1, alpha= 0.5)
plt.show()

# marker = "^" for triangle
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = [10,20,30,40,50]
label = ["Python","java","c","c++","html"]

explode = [0.0,0.8,0.0,0.0,0.0]
plt.pie(data,labels = label, explode = explode, shadow = True, colors = ["r","g","b","y","k"])
# plt.figure(figsize = (9,7))
plt.title("pie chart", fontsize = 15)
plt.show()

# explode parameter is an array
```
