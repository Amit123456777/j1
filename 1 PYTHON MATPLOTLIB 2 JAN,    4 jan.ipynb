```python
# pychart
y = np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
# pychart labels
y = np.array([35,25,25,15])
my_labels = ["apple","banana","cherry","grapes"]
plt.pie(y,labels = my_labels)
plt.show()
```

```python
# pie chart start angle parameter
y = np.array([35,25,25,15])
my_labels = ["apple","banana","cherry","grapes"]
plt.pie(y,labels = my_labels,startangle = 90)
plt.show()
```

```python
# pychart explode
y = np.array([35,25,25,15])
my_labels = ["apple","banana","cherry","grapes"]
my_explode = [0.2,0,0,0]
plt.pie(y,labels = my_labels,explode = my_explode)
plt.show()
```

```python
# pychart shadow

y = np.array([35,25,25,15])
my_labels = ["apple","banana","cherry","grapes"]
my_explode = [0.2,0,0,0]
plt.pie(y,labels = my_labels,explode = my_explode, shadow=True)
plt.show()
```

```python
# pycharm colors

# pychart labels
y = np.array([35,25,25,15])
my_labels = ["apple","banana","cherry","grapes"]
my_colors = ["black","y","hotpink","b"]
plt.pie(y,labels = my_labels,colors=my_colors)
plt.show()
```
